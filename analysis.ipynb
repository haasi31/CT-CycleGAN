# Visualize A->B Difference

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
realA = np.array(Image.open('checkpoints/syn2CT/web/images/epoch195_real_A.tiff'))
fakeB = np.array(Image.open('checkpoints/syn2CT/web/images/epoch195_fake_B.tiff'))

diff = np.abs(realA-fakeB)
```

```python
plt.figure(figsize=(20,10))
plt.subplot(131)
plt.imshow(realA, cmap='gray')
plt.axis('off')
plt.subplot(132)
plt.imshow(fakeB, cmap='gray')
plt.axis('off')
plt.subplot(133)
plt.imshow(diff, cmap='viridis')
plt.axis('off')
plt.show()
```
